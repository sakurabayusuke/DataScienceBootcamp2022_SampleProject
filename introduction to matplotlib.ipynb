# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # (width, heigh)

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("image/sample-plot.png")
```

## Making figures with Numpy arrays

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Use same data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another Scatter
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter":12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), height=nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store", ylabel="Price($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Two options for subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# plot to each different index
ax[0, 0].plot(x,x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(10));
```

### which one should you use?
* When plotting something quickly okey to use the pyplot method
* When plotting something more adbanced, use the OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
over_50.plot(kind="scatter", x="age",y="chol", c="target")
```

```python
# OO Method wtih pyplot method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter', x="age", y="chol", c="target", ax=ax);

#ax.set_xlim([45, 100])
```

```python
## OO Method from scratch
fix, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Cutomize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--")
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Desease and Chlestrerol Levels",
       ylabel="Cholesterol")

# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# add line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rage", 
        xlabel="Age",
       ylabel="Max Heart Rage")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a mealine
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
plt.style.use("seaborn-whitegrid")

## OO Method from scratch
fix, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="summer")

# Cutomize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--")
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

# Customize ax0
ax0.set(title="Heart Desease and Chlestrerol Levels",
       ylabel="Cholesterol")
ax0.set_xlim([50, 80])

# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# add line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rage", 
        xlabel="Age",
       ylabel="Max Heart Rage")

ax1.set_xlim([50, 80])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a mealine
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
